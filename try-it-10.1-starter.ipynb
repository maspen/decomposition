Then, use the `seasonal_decompose` function to decompose the series into trend and seasonality.  Examing plots of these elements, and discuss the similarities and differences between these components for the two time series.  

```python
from statsmodels.tsa.seasonal import seasonal_decompose

import statsmodels.api as sm
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# sunspots = sm.datasets.sunspots.load_pandas()
sunspots = pd.read_csv('data/Sunspots.csv')
air = pd.read_csv('data/AirPassengers.csv')
```

```python
# air = air.set_index(pd.to_datetime(air['Month']))
# air = air.drop(columns=['Month'])
# air = air.rename(columns={'#Passengers':'passengers'})
air.plot(figsize=(16,4))
```

```python
# sunspots = sunspots.set_index(pd.to_datetime(sunspots['Date']))
# sunspots = sunspots.drop(columns=['Unnamed: 0','Date'])
# sunspots = sunspots.rename(columns={'Monthly Mean Total Sunspot Number':'value'})
sunspots.plot(figsize=(16,4))
```

```python
# truncate the sunspots data to better compare w/ air
sunspots_short = sunspots['1940':'1970']

sunspots_sd = seasonal_decompose(sunspots_short, model='additive')
air_sd = seasonal_decompose(air, model='additive')
```

```python
plt.figure(figsize=(16,5))
plt.plot(sunspots_sd.observed,label='sunspots decomposed',color='orange')
plt.plot(air_sd.observed,label='air decomposed',color='blue')
plt.title('Sunspots and Air Decomposed')
plt.legend()
```

```python
plt.figure(figsize=(16,5))
plt.plot(sunspots_sd.seasonal,label='sunspots seasonal',color='orange')
plt.plot(air_sd.seasonal,label='air seasonal',color='blue')
plt.title('Sunspots and Air Seasonal')
plt.legend()
```

```python
plt.figure(figsize=(16,5))
plt.plot(sunspots_sd.trend,label='sunspots trend',color='orange')
plt.plot(air_sd.trend,label='air trend',color='blue')
plt.title('Sunspots and Air Trend')
plt.legend()
```

```python
plt.figure(figsize=(16,5))
plt.plot(sunspots_sd.resid,label='sunspots residual',color='orange')
plt.plot(air_sd.resid,label='air residual',color='blue')
plt.title('Sunspots and Air Residual Components')
plt.legend()
```

```python
plt.figure(figsize=(16,5))
plt.plot(sunspots_sd.weights,label='sunspots weights',color='orange')
plt.plot(air_sd.weights,label='air weights',color='blue')
plt.title('Sunspots and Air Weights to Reduce Outlier Influence')
plt.legend()
```
